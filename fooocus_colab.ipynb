```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sequence
n = np.arange(8)  # n from 0 to 8
x_n = (-1) ** n  # x(n) = (-1)^n

# Function to generate twiddle factor matrix
def twiddle_matrix(N):
    k = np.arange(N)
    n = k[:, None]
    return np.exp(-2j * np.pi * n * k / N)

# Compute the DFT using twiddle factors
N = len(x_n)
W = twiddle_matrix(N)  # Generate twiddle factor matrix
X_k = np.dot(W, x_n)  # Compute DFT using matrix multiplication

# Compute magnitude and phase
magnitude = np.abs(X_k)
phase = np.angle(X_k)

# Plot magnitude response
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(np.arange(N), magnitude, marker='o', linestyle='-')
plt.title('Magnitude Response')
plt.xlabel('Frequency Index (k)')
plt.ylabel('|X(k)|')
plt.grid()

# Plot phase response
plt.subplot(1, 2, 2)
plt.plot(np.arange(N), phase, marker='o', linestyle='-')
plt.title('Phase Response')
plt.xlabel('Frequency Index (k)')
plt.ylabel('Phase (radians)')
plt.grid()

plt.tight_layout()
plt.show()
```
